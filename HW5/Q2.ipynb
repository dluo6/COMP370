```python
import pandas as pd

file = '../HW4/nyc_311_limit.2020.csv'

df = pd.read_csv(file)
```

```python
df = df[df['complaint_type'] == 'Rodent']

reduced_df = df[['unique_key', 'location_type']]
grouped_df = reduced_df.groupby(['location_type'], as_index=False).count()
grouped_df.rename(columns={'unique_key': 'count'}, inplace=True)
sorted_df = grouped_df.sort_values(by='count', ascending=False)
print(sorted_df)
```

```python
num_of_location_types = 20
data_df = sorted_df.head(num_of_location_types)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

fig = plt.figure(figsize = (10, 6))

bar_fig = plt.bar(data_df['location_type'], data_df['count'])
plt.xticks(rotation=45, ha='right')

plt.xlabel('Location Types')
plt.ylabel('Count of complaints')
plt.title(f'Top {num_of_location_types} common location types with rodent complaints')
```
